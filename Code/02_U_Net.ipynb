**Image Segmentation Types**<br/>
`Semantic Segmentation:`<br/>
▪ Assigns each pixel to a specific class<br/>
▪ Doesn't distinguish between instances of same class<br/>
▪ Example: All cars are labeled as "car" class<br/>

`Instance Segmentation:`<br/>
▪ Detects and segments individual objects<br/>
▪ Distinguishes between instances of same class<br/>
▪ Example: Each car gets unique label (car1, car2)<br/>

`Panoptic Segmentation:`<br/>
▪ Combines semantic and instance segmentation<br/>
▪ Segments both countable objects and background<br/>
▪ Example: Cars as instances + road/sky as semantic<br/>

`Binary Segmentation:`<br/>
▪ Segments image into foreground and background<br/>
▪ Only two classes (0 or 1)<br/>
▪ Example: Object vs background<br/>

`Real-time Segmentation:`<br/>
▪ Fast processing for video/streaming<br/>
▪ Often trades accuracy for speed<br/>
▪ Example: Self-driving car vision systems<br/>

**Common Evaluation Metrics for Image Segmentation**<br/>
`Intersection over Union (IoU):`<br/>
▪ Measures overlap between predicted and ground truth masks<br/>
▪ Formula: (Area of Intersection) / (Area of Union)<br/>
▪ Range: 0 (no overlap) to 1 (perfect overlap)<br/>
▪ Widely used for object detection and segmentation<br/>

`Dice Coefficient / F1 Score:`<br/>
▪ Similar to IoU but gives more weight to overlapping regionsv<br/>
▪ Formula: 2 * (Area of Intersection) / (Sum of both areas)<br/>
▪ Range: 0 to 1<br/>
▪ Popular in medical image segmentation<br/>

`Pixel Accuracy:`<br/>
▪ Ratio of correctly classified pixels to total pixels<br/>
▪ Simple but can be misleading with class imbalance<br/>
▪ Formula: (True Positives + True Negatives) / Total Pixels<br/>

`Mean Pixel Accuracy:`<br/>
▪ Average per-class pixel accuracy<br/>
▪ Better for imbalanced classes<br/>
▪ Calculated separately for each class then averaged<br/>

`Mean IoU (mIoU):`<br/>
▪ Average IoU across all classes<br/>
▪ Standard metric for semantic segmentation<br/>
▪ Handles class imbalance well<br/>

`Boundary F1 Score:`<br/>
▪ Focuses on segmentation boundaries<br/>
▪ Important for precise edge detection<br/>
▪ Useful when boundary accuracy is critical<br/>

`Precision and Recall:`<br/>
▪ Precision: Accuracy of positive predictions<br/>
▪ Recall: Ability to find all positive instances<br/>
▪ Important for specific applications<br/>

Image classification assigns a single label to an entire image, while segmentation goes a step further by labeling each pixel or group of pixels, effectively dividing the image into meaningful components. This enables more detailed analysis and supports applications like object detection, medical imaging, and autonomous driving, where identifying specific parts of an image is essential.




**Segmentation methods:**<br/>
▪ `Manual annotation `involves marking image boundaries or regions of interest. Though reliable, it is time-consuming, labor-intensive, prone to errors, and unsuitable for large datasets or complex tasks due to consistency challenges.<br/>
▪ `Pixel-wise classification` independently classifies each pixel using algorithms like decision trees, SVMs, or random forests. However, it often fails to capture global context, spatial relationships, and accurate object boundaries, leading to segmentation issues.<br/>
▪ `U-Net's` architecture features a contracting path and an expansive path. The contracting path, similar to feedforward layers in other CNNs, uses encoder layers to reduce spatial resolution while capturing contextual information. The expansive path uses decoder layers to upsample feature maps and decode the data, aided by skip connections from the contracting path to preserve spatial details. This combination enables accurate feature localization and segmentation map generation.<br/>

UNET was developed to address the inefficiencies and inaccuracies of traditional image segmentation methods.<br/>
▪ `End-to-End Learning:` <br/>
UNET employs an end-to-end learning approach, enabling it to segment images directly from input-output pairs without requiring user annotations. By training on a large, labeled dataset, UNET automatically extracts essential features and delivers precise segmentation, eliminating the need for time-consuming manual labeling. <br/>

▪ `Fully Convolutional Architecture:` <br/>
UNET utilizes a fully convolutional architecture, consisting solely of convolutional layers without any fully connected layers. This design allows UNET to process input images of any size, enhancing its versatility and adaptability for diverse segmentation tasks and varying input dimensions.<br/>

▪ `U-shaped Architecture with Skip Connections:` The network's architecture features encoding and decoding paths for local and global context, with skip connections preserving key information for precise segmentation.<br/>

▪ `Contextual Information and Localisation:` Skip connections in UNET merge multi-scale features, enhancing context absorption, detail capture, and improving segmentation accuracy with precise object boundaries. <br/>

▪ `Data Augmentation and Regularization:` UNET enhances resilience and generalization by using data augmentation, like rotations, flips, scaling, and deformations, to diversify training data, and regularization techniques, such as dropout and batch normalization, to prevent overfitting. <br/>

**Key Elements of the UNET Framework:**<br/>
▪ `Contracting Path (Encoding Path):` UNET's contracting path uses convolution and max pooling to capture high-resolution, low-level features while reducing spatial dimensions.<br/>

▪ `Expanding Path (Decoding Path):` Transposed convolutions, or deconvolutions, upsample feature maps in the UNET expansion path to restore a dense segmentation map.

▪ `Skip Connections:` Skip connections in UNET link encoding and decoding layers, preserving spatial details and enhancing segmentation by merging features from earlier layers. <br/>

▪ `Concatenation:` Concatenation in UNET skip connections merges encoding path features with upsampled decoding path features, combining multi-scale information for precise segmentation. <br/>

▪ `Fully Convolutional Layers:` UNET uses convolutional layers, avoiding fully connected ones, allowing it to process images of any size while retaining spatial information for versatile segmentation tasks. <br/> <br/>

▪ `Dice Coefficient Loss:` The Dice coefficient measures overlap between predicted and true segmentation masks, while Dice loss (1 - Dice coefficient) minimizes as alignment improves. It's particularly effective for unbalanced datasets, encouraging accurate separation of foreground and background by penalizing false positives and negatives. <br/>

▪ `Cross-Entropy Loss:` It is used in image segmentation to measure dissimilarity between predicted class probabilities and ground truth labels. It treats each pixel as an independent classification problem, encouraging high probabilities for correct classes and penalizing deviations. This method works well for balanced foreground/background classes or multi-class segmentation tasks. <br/>

<font color=#036611 szie="4.5" face="Arial"><b>7️⃣ U-Net Architectur</b></font><br/>
**U-Net's architecture consists of two main paths:**<br/>

`1. Contracting Path (Encoder; Left side):`<br/>
Series of convolutional and max pooling layers (Two 3x3 convolutions (unpadded); ReLU activation; 2x2 max pooling with stride 2)<br/>
Gradually reduces spatial dimensions while increasing feature channels<br/>
Each downsampling step doubles the number of feature channels<br/>

`2. Bottleneck (Bottom):`<br/>
Bridge between contracting and expansive paths<br/>
Contains two 3x3 convolutions with ReLU<br/>
Has the highest number of feature channels<br/>

`3. Expansive Path (Decoder; Right side):`<br/>
Upsampling blocks that gradually increase spatial dimensions while reducing feature channels through:<br/>
2x2 up-convolutions with feature channel reduction<br/>
Skip connections from encoder<br/>
Double 3x3 convolutions<br/>
This enables precise localization of features<br/>

`4. Skip Connections:`<br/>
Direct connections between contracting and expansive paths<br/>
Help preserve spatial information lost during downsampling<br/>
Essential for precise localization<br/>

`5. Final Layer:`<br>
1x1 convolution to map feature vector to desired number of classes<br>
Output is a pixel-wise segmentation map<br>

`Key features:`<br/>
Works with very few training images<br/>
No fully connected layers (fully convolutional)<br/>
Preserves high resolution features through skip connections<br/>
Symmetric architecture with matching encoder-decoder levels<br/>
Final output has same resolution as input but with semantic segmentation<br/>


**Comparison of U-Net with CNN and FCN:**<br/>
**Traditional CNN vs U-Net:**<br/>
`Traditional CNN:`<br/>
- Primarily designed for classification tasks<br/>
- Uses fully connected layers at the end<br/>
- Loses spatial information through pooling<br/>
- Output is a single class label<br/>
- Not suitable for pixel-wise segmentation<br/>

`U-Net:`<br/>
- Specifically designed for segmentation<br/>
- Fully convolutional architecture<br/>
- Preserves spatial information via skip connections<br/>
- Output has same resolution as input<br/>
- Pixel-wise segmentation prediction<br/>

**FCN vs U-Net:**<br/>
`Fully Convolutional Network (FCN):`<br/>
- First architecture for end-to-end segmentation<br/>
- Uses VGG/ResNet as encoder<br/>
- Simple upsampling in decoder<br/>
- Limited skip connections<br/>
- Lower resolution output possible<br/>

`U-Net:`<br/>
- Symmetric encoder-decoder structure<br/>
- Custom encoder design<br/>
- Sophisticated decoder with skip connections<br/>
- Multiple skip connections at each level<br/>
- Maintains high resolution details<br/>
- Better for medical image segmentation<br/>

```python

# Import required libraries
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Conv2DTranspose, concatenate

def unet_model(img_size: int):
    # 1. Input Layer
    inputs = Input(img_size)
    
    # 2. Contracting Path (Encoder)
    # First block
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(inputs)
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(conv1)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)
    
    # Second block
    conv2 = Conv2D(128, 3, activation='relu', padding='same')(pool1)
    conv2 = Conv2D(128, 3, activation='relu', padding='same')(conv2)
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)
    
    # Third block
    conv3 = Conv2D(256, 3, activation='relu', padding='same')(pool2)
    conv3 = Conv2D(256, 3, activation='relu', padding='same')(conv3)
    pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)
    
    # Fourth block
    conv4 = Conv2D(512, 3, activation='relu', padding='same')(pool3)
    conv4 = Conv2D(512, 3, activation='relu', padding='same')(conv4)
    pool4 = MaxPooling2D(pool_size=(2, 2))(conv4)
    
    # 3. Bottleneck (Bridge)
    conv5 = Conv2D(1024, 3, activation='relu', padding='same')(pool4)
    conv5 = Conv2D(1024, 3, activation='relu', padding='same')(conv5)

    # 4. Expansive Path (Decoder)
    # First upsampling block
    up1 = Conv2DTranspose(512, 2, strides=(2, 2), padding='same')(conv5)
    concat1 = concatenate([conv4, up1], axis=3)
    conv6 = Conv2D(512, 3, activation='relu', padding='same')(concat1)
    conv6 = Conv2D(512, 3, activation='relu', padding='same')(conv6)
    
    # Second upsampling block
    up2 = Conv2DTranspose(256, 2, strides=(2, 2), padding='same')(conv6)
    concat2 = concatenate([conv3, up2], axis=3)
    conv7 = Conv2D(256, 3, activation='relu', padding='same')(concat2)
    conv7 = Conv2D(256, 3, activation='relu', padding='same')(conv7)
    
    # Third upsampling block
    up3 = Conv2DTranspose(128, 2, strides=(2, 2), padding='same')(conv7)
    concat3 = concatenate([conv2, up3], axis=3)
    conv8 = Conv2D(128, 3, activation='relu', padding='same')(concat3)
    conv8 = Conv2D(128, 3, activation='relu', padding='same')(conv8)
    
    # Forth upsampling block
    up4 = Conv2DTranspose(64, 2, strides=(2, 2), padding='same')(conv8)
    concat4 = concatenate([conv1, up4], axis=3)
    conv9 = Conv2D(64, 3, activation='relu', padding='same')(concat4)
    conv9 = Conv2D(64, 3, activation='relu', padding='same')(conv9)
    
    # 5. Final Layer (Output layer)
    outputs = Conv2D(1, 1, activation='sigmoid')(conv9)
    
    model = Model(inputs=inputs, outputs=outputs)
    return model

# Create the model
model = unet_model(img_size=data.shape[1:])
# model.summary()
```

```python

# Import required libraries
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Conv2DTranspose, concatenate

def unet_model(img_size: int):
    # 1. Input Layer
    inputs = Input(img_size)
    
    # 2. Contracting Path (Encoder)
    # First block
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(inputs)
    conv1 = Conv2D(64, 3, activation='relu', padding='same')(conv1)
    pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)
    
    # Second block
    conv2 = Conv2D(128, 3, activation='relu', padding='same')(pool1)
    conv2 = Conv2D(128, 3, activation='relu', padding='same')(conv2)
    pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)
    
    # Third block
    conv3 = Conv2D(256, 3, activation='relu', padding='same')(pool2)
    conv3 = Conv2D(256, 3, activation='relu', padding='same')(conv3)
    pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)
    
    
    # 3. Bottleneck (Bridge)
    conv5 = Conv2D(1024, 3, activation='relu', padding='same')(pool3)
    conv5 = Conv2D(1024, 3, activation='relu', padding='same')(conv5)

    # 4. Expansive Path (Decoder)
    # First upsampling block
    up1 = Conv2DTranspose(512, 2, strides=(2, 2), padding='same')(conv5)
    concat1 = concatenate([conv3, up1], axis=3)
    conv6 = Conv2D(512, 3, activation='relu', padding='same')(concat1)
    conv6 = Conv2D(512, 3, activation='relu', padding='same')(conv6)
    
    # Second upsampling block
    up2 = Conv2DTranspose(256, 2, strides=(2, 2), padding='same')(conv6)
    concat2 = concatenate([conv2, up2], axis=3)
    conv7 = Conv2D(256, 3, activation='relu', padding='same')(concat2)
    conv7 = Conv2D(256, 3, activation='relu', padding='same')(conv7)
    
    # Third upsampling block
    up3 = Conv2DTranspose(128, 2, strides=(2, 2), padding='same')(conv7)
    concat3 = concatenate([conv1, up3], axis=3)
    conv8 = Conv2D(128, 3, activation='relu', padding='same')(concat3)
    conv8 = Conv2D(128, 3, activation='relu', padding='same')(conv8)
    

    # 5. Final Layer (Output layer)
    outputs = Conv2D(1, 1, activation='sigmoid')(conv8)
    
    model = Model(inputs=inputs, outputs=outputs)
    return model

# Create the model
model = unet_model(img_size=data.shape[1:])
# model.summary()
```
